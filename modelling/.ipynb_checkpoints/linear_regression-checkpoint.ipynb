# Model Evaluation Function

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import time

def model_eval(y, y_pred, name=None, file=None, verbose=True, **kwargs):
    report = {}
    if name:
        report['name'] = name
        if verbose:
            print(name)
    
    report["RMSE"] = mean_squared_error(y, y_pred, squared=False)
    report["MAE"] = mean_absolute_error(y, y_pred)
    report["% <2 min"] = sum(abs(y-y_pred) < 2*60)/len(y)*100
    report["% <5 min"] = sum(abs(y-y_pred) < 5*60)/len(y)*100
    report["% <7 min"] = sum(abs(y-y_pred) < 7*60)/len(y)*100
    report["time"] = str(pd.Timestamp(round(time.time()), unit='s'))
    
    for kwarg in kwargs:
        report[kwarg] = kwargs[kwarg]
    
    if file is not None:
        with open(file, "a") as f:
            f.write(str(report)+"\n")
    if verbose:
        print(report)
    return(report)
```

# Data Preprocessing Function

```python
def gen_preprocessor(X_columns):
    ColumnTransformations = []
    num_cols = []
    cat_cols = []
    
    from sklearn.preprocessing import QuantileTransformer
    for col in X_columns:
        if 'circular' in docs[col]['type']:
            ColumnTransformations.append(
            (
                col + '_qcut_' + str(docs[col]['n_bins']),
                QuantileTransformer(n_quantiles=docs[col]['n_bins']),
                [col]
            ))
        if 'num' in docs[col]['type']:
            num_cols.append(col)
        if 'cat' in docs[col]['type']:
            cat_cols.append(col)

    from sklearn.preprocessing import StandardScaler
    from sklearn.impute import SimpleImputer
    from sklearn.pipeline import Pipeline
    num_trans = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())])
    ColumnTransformations.append(("num_trans", num_trans, num_cols))

    from sklearn.preprocessing import OneHotEncoder
    ColumnTransformations.append(("onehotencode", OneHotEncoder(handle_unknown='ignore'), cat_cols))

    from sklearn.compose import ColumnTransformer
    return ColumnTransformer(ColumnTransformations, remainder='drop', n_jobs=-1)
```

```python
import sklearn.linear_model as lm

from sklearn.pipeline import Pipeline

import time

from sklearn.base import TransformerMixin

class DenseTransformer(TransformerMixin):
    def fit(self, X, y=None, **fit_params):
        return self
    def transform(self, X, y=None, **fit_params):
        return X.toarray()

regressors = [
    lm.LinearRegression(n_jobs=-1), #
    lm.Ridge(random_state=0),#
    lm.SGDRegressor(),#
    lm.ElasticNet(),#
    lm.ElasticNetCV(n_jobs=-1),#
    lm.BayesianRidge(),#
    lm.HuberRegressor(max_iter=10000),#
    lm.RANSACRegressor(),#
    lm.PoissonRegressor(max_iter=1000, verbose=1),#
    lm.TweedieRegressor(verbose=1),#
    lm.GammaRegressor(verbose=1),#
    lm.RidgeCV(),#
    lm.Lasso(),#
    lm.OrthogonalMatchingPursuit(),#
    lm.OrthogonalMatchingPursuitCV(#
        #n_jobs=-1
        verbose=True
    ),
    lm.Lars(
        #eps = (np.finfo(float).eps) * (10)**16,
        verbose = True,
    ),#
    lm.LassoLars(
        eps = (np.finfo(float).eps) * (10)**1,
        verbose = True,
    ),#
    lm.LassoLarsIC(
        eps = (np.finfo(float).eps) * (10)**16,
        verbose = True,
    ),#
    lm.LarsCV( #
       #n_jobs=-1,
        verbose=True,
                eps = (np.finfo(float).eps) * (10)**16,
    ),
    lm.ARDRegression(verbose=True),
    lm.LassoCV(n_jobs=-1),
    lm.LassoLarsCV(
        #n_jobs=-1,
        eps = (np.finfo(float).eps) * (10)**16,
        verbose = True,
    ),
#     lm.TheilSenRegressor(n_jobs=-1, verbose=True), # FAILING
]

for i, regr in enumerate(regressors):
    regr_name = str(regr).split('(')[0]
    print(i, '/', len(regressors), regr_name)
    
    if regr_name in results['data'].keys():
        print("{} has already been sucessfully calculated".format(regr_name))
    elif regr_name in results['failures'].keys():
        print("{} previously failed, not retrying.".format(regr_name))
    elif regr_name in results['warnings'].keys():
        print("{} previously raised a warning, not retrying".format(regr_name))
    else:
        print("Calculating")
        t = time.time()
        X_cols = list(X_train.columns)

        model = Pipeline(steps=[("preprocessor", gen_preprocessor(X_cols)),
                           ("regressor", regr)])
        try:
            model.fit(X_train[X_cols], y_train)
        except Exception as inst:
            print("Error")
            print(type(inst))    # the exception instance
            print(inst)          # __str__ allows args to be printed directly,
            try:
                model = Pipeline(steps=[("preprocessor", gen_preprocessor(X_cols)),
                                        ("densifier", DenseTransformer()),
                                        ("regressor", regr)])
                model.fit(X_train[X_cols], y_train)
                print("Fix worked")
            except Exception as inst: 
                print("Fix didn't work")
                print(type(inst))    # the exception instance
                print(inst)          # __str__ allows args to be printed directly,
                continue

        results['data'][regr_name + "_eps_16"] = model_eval(
            y_val, 
            model.predict(X_val[X_cols]), 
            t=str(round(time.time()-t)),
            name="LR_varsearch_{}_h30, SIMPLE_VAL".format(regr_name),
        )
    
```

```python
from geneticalgorithm2 import geneticalgorithm2 as ga2
import sklearn.linear_model as lm

# from sklearn.utils._testing import ignore_warnings

# from sklearn.exceptions import ConvergenceWarning

regr = lm.Ridge()

# @ignore_warnings(category=ConvergenceWarning)
def model_generic(sel):
    if sum(sel)==0:
        return 0
    else:
        pen = sum(sel)*.1
        cols = np.array(X_cols)[np.array(sel, dtype=bool)].tolist()

        from sklearn.pipeline import Pipeline
        model = Pipeline(steps=[("preprocessor", gen_preprocessor(cols)),
                               ("regressor", regr)])

        model.fit(X_train[X_cols], y_train)

        #model_eval()
        # print("Finished one iteration {}".format(sum(abs(model.predict(X_test[X_columns])-y_test) < 120)/len(y_test)*100))
        return pen - model_eval(y_val, model.predict(X_val[X_cols]), verbose=False)['% <2 min']


algorithm_parameters = {'max_num_iteration': 100,\
                   'population_size':100,\
                   'mutation_probability':0.1,\
                   'elit_ratio': 0.01,\
                   'crossover_probability': 0.5,\
                   'parents_portion': 0.3,\
                   'crossover_type':'uniform',\
                   'max_iteration_without_improv':None}


ga_instance = ga2(
    model_generic,
    dimension=len(X_cols),
    function_timeout=1000,
    variable_type='bool',
    algorithm_parameters=algorithm_parameters
)

#ga_instance.run()
# 13025s or 3h 37 mins 
ga_instance.run(set_function= ga2.set_function_multiprocess(model_generic, n_jobs = -1))
```
