```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rc('font', family='Times New Roman')
```

```python
scale = 100
n_lim = 100

df = pd.read_csv("feature_importance.csv")
group_data = (np.array(df["Feature importance"])*scale).round(2).tolist()[:n_lim]
group_names = df['Column name'].to_list()[:n_lim]
```

```python
X_cols_fs = ["actype", "depgnr", "n_dep", "dew", "lightning", "rvr5000_2000", "plr1", "plr2", "local_mod", "trwy_ext"]
```

```python
fig, ax = plt.subplots(
    figsize=(8, 8)
)

ax.barh(group_names, group_data)

for i, group in enumerate(group_data):
    ax.text(group, i, group, fontsize=10,
            verticalalignment="center")
    if group_names[i] in X_cols_fs:
        ax.text(0.41*scale, i, "GA", fontsize=10,
               verticalalignment="center")


ax.set(xlim=[0, 0.4*scale], xlabel='Importance [%]', ylabel='Feature')

fig.savefig("fs_visualization.png", bbox_inches='tight')
fig.savefig("fs_visualization.pdf", format='pdf', bbox_inches='tight')

plt.show()

for group_name in X_cols_fs:
    if group_name not in group_names:
        print(group_name)

```
